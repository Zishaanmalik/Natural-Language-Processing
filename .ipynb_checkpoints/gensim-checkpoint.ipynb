```python
# Imports
import gensim
from gensim.models import Word2Vec, KeyedVectors
import gensim.downloader as api  # optional: download pretrained models
```

```python
# Small example corpus (tokenized sentences)
sentences = [
    ["i", "love", "nlp"],
    ["my", "love", "is", "nlp"],
    ["nlp", "is", "fun", "and", "powerful"],
    ["i", "am", "learning"],
    ["deep", "learning", "and", "nlp"],
    ["i", "like", "deep", "models"],
]
```

```python
# Train a tiny Word2Vec model (for learning/testing only)
model = Word2Vec(
    sentences,
    vector_size=50,   # embedding dimensionality
    window=3,         # context window
    min_count=1,      # include rare words for demo
    workers=3,        # set >1 for parallel training
    epochs=50         # more epochs for tiny corpus
)
```

```python
model.corpus_count
```

```python
# Access keyed vectors
kv = model.wv
```

```python
kv.most_similar("i")
```

```python
print("Vocab size:", len(kv.index_to_key))
print("Sample words:", kv.index_to_key[:10])
print("Vector shape for 'nlp':", kv['nlp'].shape)
```

```python
# top-5 words similar to 'nlp'
print("most_similar('nlp'):", kv.most_similar("nlp", topn=5))
```

```python
# Example of arithmetic on word vectors (analogy-style)
# Here our small model might not produce meaningful semantics, but API is same.
# positive = words to add, negative = words to subtract
try:
    print("analogy example:", kv.most_similar(positive=["learning", "love"], negative=["i"], topn=5))
except KeyError as e:
    print("Word not in vocab:", e)
```

```python
import numpy as np
vec = kv.get_vector("nlp")        # raw numpy vector for a word
print("nlp vector :\n", vec)
# compute cosine between two vectors manually
def cos(a,b): 
    return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))
print("manual cos(nlp, learning):", cos(kv.get_vector("nlp"), kv.get_vector("learning")))
```

```python
# direct meathod
kv.similarity('nlp','learning')
```

```python
kv.similarity('i','my')
```

```python
# Save only KeyedVectors (small on-disk and faster to load)

from pathlib import Path
# create folder
root = Path("models")
root.mkdir(exist_ok=True)
path = root / "small_kv.kv"

# Save KeyedVectors
kv.save(str(path)) # saves word vectors & vocabulary
           
# Load it back
kv2 = KeyedVectors.load(str(path), mmap='r')

print("Loaded kv2 vocab size:", len(kv2.index_to_key))
```

```python
# load a small pre-trained model (downloads first time)
# recommended for experiments: 'glove-wiki-gigaword-50' or 'glove-wiki-gigaword-100'
pretrained = api.load("glove-wiki-gigaword-50")   # ~50d glove
print("pretrained vocab sample:", list(pretrained.key_to_index)[:6])
print("similar to 'king':", pretrained.most_similar("king", topn=5))
```

```python
# Example: words similar to ('king' - 'man' + 'woman')
res = pretrained.most_similar(positive=["king","woman"], negative=["man"], topn=5)
print("king - man + woman =>", res)
# Multi-word average: average vector of words
def avg_vector(words, kv_obj):
    vs = [kv_obj[word] for word in words if word in kv_obj]
    return np.mean(vs, axis=0)
print("similar to average('deep','learning'):", pretrained.similar_by_vector(avg_vector(["deep","learning"], pretrained), topn=5))
```

```python
# Quick 2D visualization of few words (requires matplotlib)
# !pip install matplotlib
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

words = ["nlp","learning","deep","love","powerful","fun","i","am"]
vecs = [kv[w] for w in words if w in kv.key_to_index]
pca = PCA(n_components=2).fit_transform(vecs)

plt.figure(figsize=(6,4))
plt.scatter(pca[:,0], pca[:,1])
for i,w in enumerate(words):
    if w in kv.key_to_index:
        plt.text(pca[i,0]+0.01, pca[i,1]+0.01, w)
plt.title("PCA of word vectors (demo)")
plt.show()# Quick 2D visualization of few words (requires matplotlib)
# !pip install matplotlib
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

words = ["nlp","learning","deep","love","powerful","fun","i","am"]
vecs = [kv[w] for w in words if w in kv.key_to_index]
pca = PCA(n_components=2).fit_transform(vecs)

plt.figure(figsize=(6,4))
plt.scatter(pca[:,0], pca[:,1])
for i,w in enumerate(words):
    if w in kv.key_to_index:
        plt.text(pca[i,0]+0.01, pca[i,1]+0.01, w)
plt.title("PCA of word vectors (demo)")
plt.show()
```

```python
# similarity between two lists (sets) of words: use average vectors
A = ["deep","learning"]
B = ["natural","language","processing"]
def set_similarity(a,b, kv_obj):
    va = avg_vector(a, kv_obj)
    vb = avg_vector(b, kv_obj)
    return cos(va,vb)

print("similarity A vs B:", set_similarity(A,B, pretrained))
```
